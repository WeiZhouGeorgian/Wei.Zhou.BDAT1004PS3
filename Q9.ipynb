```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\weizh\Desktop\Moving\Georgian\Data Programming\Problem set 3\roman-emperor-reigns.csv')

assassinated = df[df['Cause_of_Death'] == 'Assassinated']
total = len(df)
assassinated_fraction = len(assassinated) / total

other_fraction = 1 - assassinated_fraction

labels = ['Assassinated', 'Other Causes']
sizes = [assassinated_fraction, other_fraction]
explode = (0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')

ax1.set_title('Fraction of Roman Emperors Assassinated')

plt.show()
```
