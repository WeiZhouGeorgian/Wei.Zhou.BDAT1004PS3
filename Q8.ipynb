```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\weizh\Desktop\Moving\Georgian\Data Programming\Problem set 3\actor_kill_counts.csv')

df.sort_values(by='Count', ascending=True, inplace=True)

ax = df.plot(kind='barh', x='Actor', y='Count', legend=False)

for i, v in enumerate(df['Count']):
    ax.text(v + 1, i, str(v), color='black', fontweight='bold')

ax.set_xlabel('Kill Count')
ax.set_ylabel('Actor')
ax.set_title('Deadliest Actors in Hollywood')

plt.show()
```
