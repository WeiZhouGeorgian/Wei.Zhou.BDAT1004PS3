```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\weizh\Desktop\Moving\Georgian\Data Programming\Problem set 3\arcade-revenue-vs-cs-doctorates.csv')

df['Year'] = pd.to_numeric(df['Year'])

fig, ax = plt.subplots()
sc = ax.scatter(df['Total Arcade Revenue (billions)'], df['Computer Science Doctorates Awarded (US)'], c=df['Year'], cmap='viridis')

ax.set_xlabel('Total Arcade Revenue (billions)')
ax.set_ylabel('Computer Science Doctorates Awarded (US)')
ax.set_title('Arcade Revenue vs CS Doctorates (2000-2009)')

cbar = plt.colorbar(sc)
cbar.set_label('Year')

plt.show()
```
